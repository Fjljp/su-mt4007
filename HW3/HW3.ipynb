# Visualisation of IRIS data

```python
import pandas as pd

data = pd.read_csv('IRIS.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
# y=x = petal_width
# y= petal_length, x= sepal_length

# split data into 3
setosa=data[data["species"]=="Iris-setosa"]
versi=data[data["species"]=="Iris-versicolor"]
virg=data[data["species"]=="Iris-virginica"]

# create plot
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(setosa["sepal_length"],setosa["petal_length"], color='blue', label='Iris-setosa')
plt.scatter(versi["sepal_length"],versi["petal_length"], color='green', label='Iris-versicolor')
plt.scatter(virg["sepal_length"],virg["petal_length"], color='red', label='Iris-virginica')
# plt.title('Scatter Plot showing sepal length plotted against petal length of 3 different Iris flowers')
plt.grid(True)
plt.xlabel('Sepal_length')
plt.ylabel('petal_length')

plt.subplot(1, 2, 2)
plt.scatter(setosa["sepal_width"],setosa["petal_width"], color='blue', label='Iris-setosa')
plt.scatter(versi["sepal_width"],versi["petal_width"], color='green', label='Iris-versicolor')
plt.scatter(virg["sepal_width"],virg["petal_width"], color='red', label='Iris-virginica')
plt.grid(True)
plt.xlabel('Sepal_width')
plt.ylabel('petal_width')

plt.tight_layout()
plt.legend()

plt.legend(loc='center right')
plt.show
```

## As can be observed in the figures above, there is a clear linear separation between different flower specicies, based on their petal and sepal width, as well as their petal and sepal length.


```python
import matplotlib.pyplot as plt
import pandas as pd
# split data
setosa=data[data["species"]=="Iris-setosa"]
versi=data[data["species"]=="Iris-versicolor"]
virg=data[data["species"]=="Iris-virginica"]


# Boxplots
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
bplot1=plt.boxplot(
    [setosa["sepal_width"], versi["sepal_width"], virg["sepal_width"]],
    labels=["Setosa", "Versicolor", "Virginica"],
    patch_artist=True  
)
colors = ['blue', 'green', 'red']
for patch, color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)


plt.title("Sepal Width")
plt.ylabel("Width")


plt.subplot(2, 2, 2)
bplot2=plt.boxplot(
    [setosa["sepal_length"], versi["sepal_length"], virg["sepal_length"]],
    labels=["Setosa", "Versicolor", "Virginica"],
    patch_artist=True
)

for patch, color in zip(bplot2['boxes'], colors): 
    patch.set_facecolor(color)

plt.title("Sepal Length")
plt.ylabel("Length")


plt.subplot(2, 2, 3)
bplot3=plt.boxplot(
    [setosa["petal_width"], versi["petal_width"], virg["petal_width"]],
    labels=["Setosa", "Versicolor", "Virginica"],
    patch_artist=True
)
for patch, color in zip(bplot3['boxes'], colors):
    patch.set_facecolor(color)
    
plt.title("Petal Width")
plt.ylabel("Width")


plt.subplot(2, 2, 4)
bplot4=plt.boxplot(
    [setosa["petal_length"], versi["petal_length"], virg["petal_length"]],
    labels=["Setosa", "Versicolor", "Virginica"],
    patch_artist=True
)
for patch, color in zip(bplot4['boxes'], colors):
    patch.set_facecolor(color)
    
plt.title("Petal Length")
plt.ylabel("Length")

plt.show()


```

## Here I can conclude that the Setosa species seperates well from the other two by a lower median for the petal width, length and sepal length, but higher for sepal width. Additionally, there is a more overlapping separation but sitll prominent difference between verisoclour and virginica for all measurements. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
# specify characteristics and plot a pairsplot using seaborn
sns.pairplot(data, diag_kind="kde", hue="species", palette=["blue", "green", "red"],
             plot_kws={"alpha": 0.2})
plt.show()


```

## Here we can see that sepal length is correlated with sepal width for Versicolour and Virginica, but not as strong within Iris-Setosa. There is a strong correlation between petal length and sepal length for Versicolour and Virginica, but no significant correlation observed wihtin Setosa. Similarly, for the sepal length and petal width there is correlation for Versicolour and Virginica but no clear correlation within Setosa.  

# Birdwatching

```python
import pandas as pd

# retrieve file
datab = pd.read_csv('artportalen.csv')
datab.head()
```

```python
import pandas as pd

# Remove all missing values 
datab['Antal'] = pd.to_numeric(datab['Antal'], errors='coerce')
db = datab[['Artnamn', 'Antal']].dropna(subset=['Antal'])


# show counts of all species 
species_count = db.groupby('Artnamn')['Antal'].sum()
sorted_v=species_count.sort_values(ascending=False)

# The most prevalent species:
sorted_v.head(1)
```

```python
 print(sorted_v.head(3))
# The rarest species:
sorted_v.tail(2)
```

```python
# The top 3 most prevalent species: 
sorted_v.head(3)
```

```python
# Distribution of top 3 most prevalent species

top3 = ["Grönsiska", "Sothöna", "Gräsand"]

# remove missing values and filter to 4 columns
db2= datab[datab['Artnamn'].isin(top3)][['Artnamn', 'Antal', 
# split data into groups        
                                         #'Startdatum', 'Slutdatum']].dropna(subset=['Antal'])
Grön=db2[db2['Artnamn']=='Grönsiska']
Sot=db2[db2['Artnamn']=='Sothöna']
Gräs=db2[db2['Artnamn']=='Gräsand']
# Grön_count = Grön.groupby(['Artnamn','Startdatum'])['Antal'].sum() # , 'Slutdatum'
# Sot_count = Sot.groupby(['Artnamn','Startdatum'])['Antal'].sum()
# Gräs_count = Gräs.groupby(['Artnamn','Startdatum'])['Antal'].sum()

# sort by start date and include count

Grön_count = Grön.groupby('Startdatum')['Antal'].sum().reset_index()  
Sot_count = Sot.groupby('Startdatum')['Antal'].sum().reset_index()
Gräs_count = Gräs.groupby('Startdatum')['Antal'].sum().reset_index()

# merge groups to one dataframe 
dis = pd.merge(Grön_count, Sot_count, on='Startdatum', how='outer')
dis = pd.merge(dis, Gräs_count, on='Startdatum', how='outer')

# name the columns for clarity 
#dis = pd.concat([Grön_count, Sot_count, Gräs_count], axis=1)
dis.columns = ['Startdatum','Grön_count', 'Sot_count', 'Gräs_count']
# show all rows
pd.set_option('display.max_rows', None)

print(dis)
```
